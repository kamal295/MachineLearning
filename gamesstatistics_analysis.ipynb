Top_10s and Wins: A strong positive correlation suggests that players who often finish in the top 10 also tend to win more matches.

There are other correlations that can be inferred from the heatmap, but these are some of the prominent ones.

```python
#Applying Feature Engineering and create Kill-Death Ratio (KDR), Win Rate, and Average Damage Per Match features.
# Kill-Death Ratio (KDR)
df['KDR'] = df['Kills'] / df['Deaths']

# Win Rate
df['Win_Rate'] = (df['Wins'] / df['Matches_Played']) * 100

# Average Damage Per Match
df['Avg_Damage'] = df['Damage_Dealt'] / df['Matches_Played']

# Displaying the first few rows with the new features
df[['Player_Name', 'KDR', 'Win_Rate', 'Avg_Damage']].head()
```

KDR (Kill-Death Ratio): It represents how many kills a player gets for each death. A player with a KDR of 4, for instance, would get an average of 4 kills before dying.

Win Rate: This is the percentage of matches a player has won out of all the matches they've played.

Average Damage Per Match: This metric represents the average damage a player deals in a single match.

```python
# Plotting the distributions of the new features

# Initializing the figure
plt.figure(figsize=(15, 5))

# List of new features
new_features = ['KDR', 'Win_Rate', 'Avg_Damage']

# Creating subplots for each new feature
for i, feature in enumerate(new_features, 1):
    plt.subplot(1, 3, i)
    sns.histplot(df[feature], bins=20, kde=True, color='purple')
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

The distributions of the newly engineered features are as follows:

KDR (Kill-Death Ratio): The distribution is right-skewed, indicating that most players have a KDR between 3.5 and 5. This suggests that many players, on average, achieve 3 to 5 kills before they are killed in a match.

Win Rate: The distribution shows that most players have a win rate between 13% and 15%. This means that many players win roughly 13 to 15 out of every 100 matches they play.

Average Damage Per Match: The distribution is somewhat normal, centered around 60. Most players deal an average damage of around 60 per match.
